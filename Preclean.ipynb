```python
import pandas as pd

df = pd.read_csv("charging.csv")
df.head()
```

```python
df
```

```python
print(df.columns)
```

```python
# now we want to drop the un-used columns
df = df.drop(['Time of Day', 'Day of Week', 'Charging Station ID'], axis=1)
```

```python
df
```

```python
# now we want to handle the na
missing_values = df.isna().any()
print(missing_values)

# from the missing value we can see there are 3 columns has missing values
import matplotlib.pyplot as plt
import seaborn as sns

#  Energy Consumed
sns.histplot(df['Energy Consumed (kWh)'].dropna(), kde=True)
plt.title('Energy Consumed Distribution')
plt.show()

# VCharging Rate
sns.boxplot(x=df['Charging Rate (kW)'])
plt.title('Charging Rate Boxplot')
plt.show()

#driving rate
sns.boxplot(x=df['Distance Driven (since last charge) (km)'])
plt.title('Driving Rate Boxplot')
plt.show()

# Based on the plots we will use mean to fill the na
mean_energy = df['Energy Consumed (kWh)'].mean()
df['Energy Consumed (kWh)'].fillna(mean_energy, inplace=True)


mean_charging_rate = df['Charging Rate (kW)'].mean()
df['Charging Rate (kW)'].fillna(mean_charging_rate, inplace=True)

mean_distance = df['Distance Driven (since last charge) (km)'].mean()
df['Distance Driven (since last charge) (km)'].fillna(mean_distance, inplace=True)
```

```python
# we want to calculate the duration
df['Charging Start Time'] = pd.to_datetime(df['Charging Start Time'])
df['Charging End Time'] = pd.to_datetime(df['Charging End Time'])

# Calculate the time difference and create a new column
df['Charging Duration'] = df['Charging End Time'] - df['Charging Start Time']

# duration in minutes
df['Charging Duration (Minutes)'] = df['Charging Duration'].dt.total_seconds() / 60


```

```python
df
```

```python
# we want to calculate the cost per minute
df['charging cost per min'] = df['Charging Cost (USD)'] / df['Charging Duration (Minutes)']
```

```python
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df, train_size=0.6, test_size=0.4, random_state=42)

df.to_csv('new_charging.csv', index=False)
train_df.to_csv('train.csv', index=False)
test_df.to_csv('test.csv', index=False)
```
